```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ankhang_df = pd.read_csv(r'D:\OU\KhoaLuanTotNghiep\DrugRecommandation\crawlData_AnKhang\ankhang_medicines_data_full.csv', encoding='utf-8-sig')
print(f"Số dòng: {ankhang_df.shape[0]}, Số cột: {ankhang_df.shape[1]}")
ankhang_df.info()
ankhang_df.head()
```

```python
# Kiểm tra các dòng còn thiếu category
category_missing = ankhang_df[ankhang_df['category'].isnull()]
print(f"Số dòng thiếu category: {category_missing.shape[0]}")
category_missing.head()
```

```python
# Kiểm tra số dòng có trùng url
duplicate_urls = ankhang_df[ankhang_df.duplicated(subset=['url'], keep=False)]
print(f"Số dòng có trùng url: {duplicate_urls.shape[0]}")
duplicate_urls.head()
```

```python
# Loại bỏ những dòng có lỗi (cột error có dữ liệu)
error_rows = ankhang_df[ankhang_df['error'].notnull()]
print(f"Số dòng có lỗi: {error_rows.shape[0]}")
error_rows.head()
```

```python
# Loại bỏ những dòng có lỗi
ankhang_df_cleaned = ankhang_df[ankhang_df['error'].isnull()]
print(f"Số dòng sau khi loại bỏ lỗi: {ankhang_df_cleaned.shape[0]}")
ankhang_df_cleaned.head()
```

```python
# Hoặc xem chi tiết một url cụ thể
sample_url = duplicate_urls['url'].iloc[1]
ankhang_df_cleaned[ankhang_df_cleaned['url'] == sample_url]
```

```python
# Giữ dòng đầu tiên của mỗi url trùng lặp
ankhang_df_cleaned = ankhang_df_cleaned.drop_duplicates(subset=['url'], keep='first')
print(f"Số dòng sau khi loại bỏ trùng url: {ankhang_df_cleaned.shape[0]}")
```

```python
ankhang_df_cleaned.info()
ankhang_df_cleaned.head()
```

```python
# Loại bỏ cột error
ankhang_df_cleaned = ankhang_df_cleaned.drop(columns=['error'])
print(f"Số dòng sau khi loại bỏ cột error: {ankhang_df_cleaned.shape[0]}, Số cột: {ankhang_df_cleaned.shape[1]}")
ankhang_df_cleaned.info()
```

```python
ankhang_df_link = pd.read_csv(r'D:\OU\KhoaLuanTotNghiep\DrugRecommandation\crawlData_AnKhang\ankhang_medicine_links.csv', encoding='utf-8-sig')
ankhang_df_link.info()
ankhang_df_link.head()
```

```python
ankhang_df_final = pd.merge(
    ankhang_df_cleaned,
    ankhang_df_link[['link', 'category', 'drug_type']].drop_duplicates('link'),
    left_on='url',
    right_on='link',
    how='left',
    validate='m:1'
).drop(columns=['link'])  # bỏ cột links sau khi ghép nếu không cần

print(f"Số dòng sau khi merge: {ankhang_df_final.shape[0]}, Số cột: {ankhang_df_final.shape[1]}")
ankhang_df_final.info()
```

```python
print(f"category_x non-null: {ankhang_df_final['category_x'].notna().sum()}")
print(f"category_y non-null: {ankhang_df_final['category_y'].notna().sum()}")

# Kiểm tra các dòng có cả 2 giá trị
both_not_null = ankhang_df_final[ankhang_df_final['category_x'].notna() & ankhang_df_final['category_y'].notna()]
print(f"\nSố dòng có cả category_x và category_y: {len(both_not_null)}")

# So sánh giá trị khi cả 2 đều có dữ liệu
same_values = both_not_null[both_not_null['category_x'] == both_not_null['category_y']]
diff_values = both_not_null[both_not_null['category_x'] != both_not_null['category_y']]
print(f"Số dòng giống nhau: {len(same_values)}")
print(f"Số dòng KHÁC NHAU: {len(diff_values)}")
```

```python
print(f"drug_type_x non-null: {ankhang_df_final['drug_type_x'].notna().sum()}")
print(f"drug_type_y non-null: {ankhang_df_final['drug_type_y'].notna().sum()}")

both_not_null_dt = ankhang_df_final[ankhang_df_final['drug_type_x'].notna() & ankhang_df_final['drug_type_y'].notna()]
same_values_dt = both_not_null_dt[both_not_null_dt['drug_type_x'] == both_not_null_dt['drug_type_y']]
diff_values_dt = both_not_null_dt[both_not_null_dt['drug_type_x'] != both_not_null_dt['drug_type_y']]

print(f"Số dòng giống nhau: {len(same_values_dt)}")
print(f"Số dòng KHÁC NHAU: {len(diff_values_dt)}")

if len(diff_values_dt) > 0:
    print(diff_values_dt[['ten_thuoc', 'drug_type_x', 'drug_type_y']].head(10))
```

```python
ankhang_df_fix = ankhang_df_final.drop(columns=['category_x', 'drug_type_x'])
ankhang_df_fix = ankhang_df_fix.rename(columns={'category_y': 'category', 'drug_type_y': 'drug_type'})
ankhang_df_fix.info()
ankhang_df_fix.head()
```

```python
ankhang_df_fix.to_csv("Ankhang_data_preprocessed.csv", index=False, encoding='utf-8-sig')
```

```python
ankhang_df = pd.read_csv("Ankhang_data_preprocessed.csv", encoding='utf-8-sig')
print(ankhang_df.shape)
ankhang_df.info()
ankhang_df.head()
```

```python
# Lấy 1 thuốc bất kỳ để kiểm tra thông tin 'công dụng tóm tắt' và 'công dụng chi tiết'
sample_medicine = ankhang_df['ten_thuoc'].iloc[2]
sample_row = ankhang_df[ankhang_df['ten_thuoc'] == sample_medicine].iloc[0]

for col in ['ten_thuoc', 'cong_dung_tom_tat', 'cong_dung_chi_tiet']:
    print(f"\n{'='*50}")
    print(f"{col}:")
    print(sample_row[col])
```

```python
# Kiểm tra số lượng thiếu công dụng tóm tắt và công dụng chi tiết
missing_summary = ankhang_df['cong_dung_tom_tat'].isnull().sum()
missing_detail = ankhang_df['cong_dung_chi_tiet'].isnull().sum()
print(f"Số dòng thiếu công dụng tóm tắt: {missing_summary}")
print(f"Số dòng thiếu công dụng chi tiết: {missing_detail}")
```

```python
# Tạo cột mới để thống nhất công dụng
ankhang_df['cong_dung'] = ankhang_df.apply(
    lambda row: row['cong_dung_chi_tiet'] if pd.notnull(row['cong_dung_chi_tiet']) else row['cong_dung_tom_tat'],
    axis=1
)
```
