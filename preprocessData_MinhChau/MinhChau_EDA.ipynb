```python
import pandas as pd
import numpy as np
```

```python
minhchau_df = pd.read_csv(r'D:\OU\KhoaLuanTotNghiep\DrugRecommandation\crawlData_MinhChau\minhchau_medicines_data.csv')
minhchau_df.info()
minhchau_df.head()
```

```python
# Kiểm tra các dòng có lỗi (cột error có giá trị)
error_rows = minhchau_df[minhchau_df['error'].notna()]
# Hiển thị các dòng có lỗi
print(error_rows[['name', 'error']])
```

```python
# Đếm số dòng có lỗi unknown error (có chứa giá trị 'unknown error: ' trong cột error)
unknown_error_count = error_rows[error_rows['error'].str.contains('unknown error: ', case=False, na=False)].shape[0]
print(f"Số dòng có lỗi 'unknown error: ': {unknown_error_count}")
print("In ra các dòng có lỗi 'unknown error: ':")
unknown_error_rows = error_rows[error_rows['error'].str.contains('unknown error: ', case=False, na=False)]
unknown_error_rows[['name', 'error']]
```

```python
# Kiểm tra tên thuốc có trùng nhau không
duplicate_names = minhchau_df[minhchau_df.duplicated(subset=['name'], keep=False)]
print(f"Số dòng có tên thuốc trùng nhau: {duplicate_names.shape[0]}")
print("In ra các dòng có tên thuốc trùng nhau:")
duplicate_names[['name', 'error']]
```

```python
# Kiểm tra có bao nhiêu tên thuốc bị NaN
nan_name_count = minhchau_df['name'].isna().sum()
print(f"Số dòng có tên thuốc bị NaN: {nan_name_count}")
```

```python
# Kiểm tra các hàng có thông tin cột Name,link, category nhưng các cột khác đều bị NaN
non_na_col = ['name', 'link', 'category','images']
other_cols = minhchau_df.columns.difference(non_na_col)

mask = (
    minhchau_df[non_na_col].notna().all(axis=1) &
    minhchau_df[other_cols].isna().all(axis=1)
)

rows_only_core = minhchau_df[mask]
rows_only_core.shape[0]
rows_only_core 
```

```python
# Số tên thuốc trùng khác nan
non_nan_duplicate_names = duplicate_names[duplicate_names['name'].notna()]
print(f"Số dòng có tên thuốc trùng nhau nhưng không phải NaN: {non_nan_duplicate_names.shape[0]}")
print("In ra các dòng có tên thuốc trùng nhau nhưng không phải NaN:")
non_nan_duplicate_names[['name']]
```

```python
# Kiểm tra số lần trùng của mỗi tên thuốc trong non_nan_duplicate_names
duplicate_name_counts = non_nan_duplicate_names['name'].value_counts()
print("Số lần trùng của mỗi tên thuốc trong non_nan_duplicate_names:")
print(duplicate_name_counts)
```

```python
# Cột khóa
key_col = 'name'
non_key_cols = non_nan_duplicate_names.columns.difference([key_col])

# Với mỗi name, kiểm tra xem có cột nào có >1 giá trị khác nhau
diff_mask = (
    non_nan_duplicate_names
    .groupby(key_col)[non_key_cols]
    .nunique(dropna=False)  # coi NaN là một giá trị để so sánh
    .gt(1)
    .any(axis=1)
)

duplicate_name_groups = diff_mask[diff_mask].index

print(f"Số tên thuốc trùng nhau nhưng nội dung khác nhau: {len(duplicate_name_groups)}")
print("In ra các tên thuốc trùng nhau nhưng nội dung khác nhau:")
print(duplicate_name_groups)
```

```python
# Bảng số lượng giá trị khác nhau theo từng cột
nunique_table = (
    non_nan_duplicate_names
    .groupby(key_col)[non_key_cols]
    .nunique(dropna=False)  # NaN cũng tính là một giá trị
)

# Chỉ giữ những cột có >1 giá trị (tức là khác nhau)
diff_cols_per_name = nunique_table.gt(1)

# In ra các cột khác nhau theo từng name
for name, row in diff_cols_per_name.iterrows():
    diff_cols = row[row].index.tolist()
    if diff_cols:
        print(f"{name} khác ở cột: {diff_cols}")
```

```python
# Lấy 1 trường hợp có tên thuốc trùng nhau nhưng nội dung khác nhau để xem chi tiết
example_name = duplicate_name_groups[1]
example_rows = non_nan_duplicate_names[non_nan_duplicate_names['name'] == example_name]
print(f"Chi tiết các dòng có tên thuốc '{example_name}' trùng nhau nhưng nội dung khác nhau:")
example_rows
```
