```python
import pandas as pd
import numpy as np
import re
import unicodedata
```

```python
minhchau_df = pd.read_csv(r'D:\OU\KhoaLuanTotNghiep\DrugRecommandation\crawlData_MinhChau\minhchau_medicines_data.csv')
cleaned_df = minhchau_df.copy()
print(cleaned_df.shape)
cleaned_df.info()
cleaned_df.head()
```

```python
# Loại bỏ cột trống hoàn toàn
cleaned_df = cleaned_df.dropna(axis=1, how='all')
print(f"Sau khi loại bỏ cột trống hoàn toàn{cleaned_df.shape}")
```

```python
missing_token = {"", "nan", "none", "null", "na", "n/a", "không có", "-"}
def norm_text(x):
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    s = unicodedata.normalize("NFC", s)
    s = re.sub(r"\s+", " ", s)  
    return s

def to_missing(x):
    if pd.isna(x):
        return np.nan
    s = str(x).strip().lower()
    return np.nan if s in missing_token else x

# Chuẩn hóa toàn bộ cột object
obj_cols = cleaned_df.select_dtypes(include=["object"]).columns
for c in obj_cols:
    cleaned_df[c] = cleaned_df[c].map(norm_text).map(to_missing)
```

```python
if "error" not in cleaned_df.columns:
    cleaned_df["error"] = np.nan

cleaned_df["has_error"] = cleaned_df["error"].notna()
cleaned_df["is_unknown_error"] = cleaned_df["error"].str.contains("unknown error", case=False, na=False)
cleaned_df["error_type"] = cleaned_df["error"].str.extract(r"^([A-Za-z_]+Error|[A-Za-z_]+Exception)", expand=False)

quarantine_df = cleaned_df[cleaned_df["has_error"]].copy()
work_df = cleaned_df[~cleaned_df["has_error"]].copy()

print("raw_df:", minhchau_df.shape)
print("clean_df:", cleaned_df.shape)
print("work_df (không lỗi):", work_df.shape)
print("quarantine_df (có lỗi):", quarantine_df.shape)
```

```python
# Các cột hay cần union
UNION_PIPE_COLS = [c for c in ["images", "category"] if c in work_df.columns]

def is_empty(v):
    return pd.isna(v) or str(v).strip() == ""

def pick_longest(series):
    vals = [str(v).strip() for v in series if not is_empty(v)]
    if not vals:
        return np.nan
    # Ưu tiên chuỗi dài nhất (thường nhiều thông tin hơn)
    return max(vals, key=len)

def union_pipe(series):
    vals = []
    for v in series.dropna():
        parts = [p.strip() for p in str(v).split("|") if p.strip()]
        vals.extend(parts)
    if not vals:
        return np.nan
    # giữ thứ tự xuất hiện
    uniq = list(dict.fromkeys(vals))
    return " | ".join(uniq)

def merge_group(g):
    out = {}
    for c in g.columns:
        if c in UNION_PIPE_COLS:
            out[c] = union_pipe(g[c])
        elif c == "error":
            # trong work_df thường rỗng, vẫn xử lý an toàn
            non_empty = [str(v).strip() for v in g[c] if not is_empty(v)]
            out[c] = " || ".join(dict.fromkeys(non_empty)) if non_empty else np.nan
        else:
            out[c] = pick_longest(g[c])
    out["n_rows_merged_link"] = len(g)
    return pd.Series(out)

if "link" in work_df.columns:
    dedup_link_df = work_df.groupby("link", dropna=False, as_index=False).apply(merge_group).reset_index(drop=True)
else:
    dedup_link_df = work_df.copy()
    dedup_link_df["n_rows_merged_link"] = 1

print("Sau gộp theo link:", dedup_link_df.shape)
```

```python
def normalize_name(s):
    if pd.isna(s):
        return np.nan
    x = str(s).strip().lower()
    x = unicodedata.normalize("NFC", x)
    x = re.sub(r"\s+", " ", x)
    return x

if "name" in dedup_link_df.columns:
    name_stage = dedup_link_df.copy()
    name_stage["name_norm"] = name_stage["name"].map(normalize_name)

    # chỉ gộp những dòng có name_norm; dòng name trống giữ nguyên
    has_name = name_stage["name_norm"].notna()
    a = name_stage[has_name].copy()
    b = name_stage[~has_name].copy()

    dedup_name_df = (
        a.groupby("name_norm", dropna=False, as_index=False)
         .apply(merge_group)
         .reset_index(drop=True)
    )

    # bỏ name_norm phụ
    if "name_norm" in dedup_name_df.columns:
        dedup_name_df = dedup_name_df.drop(columns=["name_norm"], errors="ignore")

    final_df = pd.concat([dedup_name_df, b.drop(columns=["name_norm"], errors="ignore")], ignore_index=True)
else:
    final_df = dedup_link_df.copy()

print("Sau gộp theo name:", final_df.shape)
```

```python
def contains_any(text, keywords):
    if pd.isna(text):
        return False
    t = str(text).lower()
    return any(k in t for k in keywords)

# Cờ nghi ngờ: chống chỉ định bị rơi vào chỉ định
if {"cong_dung_chi_dinh", "chong_chi_dinh"}.issubset(final_df.columns):
    final_df["contra_suspect"] = (
        final_df["chong_chi_dinh"].isna()
        & final_df["cong_dung_chi_dinh"].map(lambda x: contains_any(x, ["chống chỉ định", "không sử dụng", "không dùng", "cấm dùng"]))
    )
else:
    final_df["contra_suspect"] = False

# Cờ nghi ngờ: tương tác thuốc bị thiếu
if {"cong_dung_chi_dinh", "tuong_tac_thuoc"}.issubset(final_df.columns):
    final_df["interaction_suspect"] = (
        final_df["tuong_tac_thuoc"].isna()
        & final_df["cong_dung_chi_dinh"].map(lambda x: contains_any(x, ["tương tác", "dùng chung", "kết hợp thuốc"]))
    )
else:
    final_df["interaction_suspect"] = False

important_cols = [c for c in [
    "name", "category", "link", "images",
    "thanh_phan", "cong_dung_chi_dinh", "chong_chi_dinh",
    "tac_dung_phu", "tuong_tac_thuoc", "Bao_quan", "error"
] if c in final_df.columns]

report = pd.DataFrame({
    "column": important_cols,
    "non_null": [final_df[c].notna().sum() for c in important_cols],
    "null": [final_df[c].isna().sum() for c in important_cols],
    "fill_rate_%": [round(final_df[c].notna().mean() * 100, 2) for c in important_cols]
}).sort_values("fill_rate_%", ascending=False)

print("=== Quality report ===")
display(report)

print("\nSố dòng contra_suspect:", int(final_df["contra_suspect"].sum()))
print("Số dòng interaction_suspect:", int(final_df["interaction_suspect"].sum()))
```

```python
def pick_by_kw(text, kws):
    if pd.isna(text): 
        return pd.NA
    s = str(text)
    low = s.lower()
    hits = [low.find(k) for k in kws if k in low]
    if not hits:
        return pd.NA
    i = min(hits)
    return s[i:].strip()

# fill chong_chi_dinh nếu đang rỗng
m1 = final_df["chong_chi_dinh"].isna()
final_df.loc[m1, "chong_chi_dinh"] = final_df.loc[m1, "cong_dung_chi_dinh"].map(
    lambda x: pick_by_kw(x, ["chống chỉ định", "không sử dụng", "không dùng", "cấm dùng"])
)

# fill tuong_tac_thuoc nếu đang rỗng
m2 = final_df["tuong_tac_thuoc"].isna()
final_df.loc[m2, "tuong_tac_thuoc"] = final_df.loc[m2, "cong_dung_chi_dinh"].map(
    lambda x: pick_by_kw(x, ["tương tác", "dùng chung", "kết hợp thuốc"])
)

# cập nhật lại suspect
final_df["contra_suspect"] = final_df["chong_chi_dinh"].isna() & final_df["cong_dung_chi_dinh"].str.contains(
    "chống chỉ định|không sử dụng|không dùng|cấm dùng", case=False, na=False
)
final_df["interaction_suspect"] = final_df["tuong_tac_thuoc"].isna() & final_df["cong_dung_chi_dinh"].str.contains(
    "tương tác|dùng chung|kết hợp thuốc", case=False, na=False
)

print("contra_suspect:", int(final_df["contra_suspect"].sum()))
print("interaction_suspect:", int(final_df["interaction_suspect"].sum()))
```
