```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pharmacity_df = pd.read_csv("pharmacity_medicines_data_final.csv")
print(pharmacity_df.shape)
pharmacity_df.head()
```

```python
# Các cột hiện có trong data
print("Các cột hiện có trong data:")
print(pharmacity_df.columns)
print("Các cột có giá trị thiếu:")
print(pharmacity_df.isnull().sum())
```

```python
pharmacity_df = pharmacity_df.copy()

# Loại bỏ các cột không cần thiết hoặc đã được xử lý
columns_to_drop = ['error', 'category', 'category_link','thong_tin_san_xuat']
pharmacity_df = pharmacity_df.drop(columns=columns_to_drop)
print("Số cột còn lại:")
print(pharmacity_df.shape[1])
print("Các cột sau khi loại bỏ:")
print(pharmacity_df.columns)
```

```python
# Thay NaN bằng chuỗi rỗng trước khi gộp
pharmacity_df['huong_dan_su_dung'] = (
    pharmacity_df['huong_dan_su_dung'].fillna('') + ' ' + 
    pharmacity_df['cach_su_dung'].fillna('')
).str.strip()  # Loại bỏ khoảng trắng dư thừa

# Loại bỏ cột 'cách sử dụng' sau khi gộp
pharmacity_df = pharmacity_df.drop(columns=['cach_su_dung'])

# Kiểm tra
print(f"NaN sau khi gộp: {pharmacity_df['huong_dan_su_dung'].isna().sum()}")
print(f"Chuỗi rỗng: {(pharmacity_df['huong_dan_su_dung'] == '').sum()}")
pharmacity_df['huong_dan_su_dung'].head(10)
```

```python
pharmacity_df.to_csv("pharmacity_medicines_data_final_cleaned.csv", index=False)
```
