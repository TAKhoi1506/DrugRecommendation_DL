```python
import pandas as pd
import numpy as np
```

```python
df_fixed = pd.read_csv("pharmacity_medicines_recrawled_fixed.csv")
print(df_fixed.shape)
df_fixed.head()
```

```python
df_base = pd.read_csv(r"D:\OU\KhoaLuanTotNghiep\DrugRecommandation\crawlData_Pharmacity\pharmacity_medicines_data_demo(9).csv")
print(df_base.shape)
df_base.head()
```

```python
# Chuẩn hóa url
for df in [df_fixed, df_base]:
    df['url'] = df['url'].astype(str).str.strip()  
```

```python
# Loại trùng
df_fix = df_fixed.drop_duplicates(subset='url',keep='last')
```

```python
# Xác định các url có lỗi trong df_base
name_col = df_base['name'].astype(str).str.strip().str.lower()
error_mask = df_base['name'].isna() | name_col.isin(['none', 'nan', ''])
error_urls = set(df_base.loc[error_mask, 'url'])
```

```python
df_fix_target = df_fix[df_fix['url'].isin(error_urls)].copy()
base_idx = df_base.set_index('url')
fix_idx = df_fix_target.set_index('url')
```

```python
replace_cols = [c for c in fix_idx.columns if c in base_idx.columns]
match_urls = base_idx.index.intersection(fix_idx.index)
```

```python
base_idx.loc[match_urls, replace_cols] = fix_idx.loc[match_urls, replace_cols]
df_merged = base_idx.reset_index()
df_merged.to_csv("pharmacity_medicines_data_merged.csv", index=False)
```

```python
new_name_col = df_merged['name'].astype(str).str.strip().str.lower()
remaining_error = df_merged['name'].isna() | new_name_col.isin(['none', 'nan', ''])

print(f"Tổng dòng gốc: {len(df_base)}")
print(f"Số URL lỗi ban đầu: {len(error_urls)}")
print(f"Số URL fix match và đã thay: {len(match_urls)}")
print(f"Số dòng còn lỗi sau thay: {remaining_error.sum()}")
print(f"Đã lưu: pharmacity_medicines_data_merged.csv")
```

```python
# Tìm số dòng còn thiếu category
missing_category = df_merged['category_name'].isna().sum()
print(f"Số dòng còn thiếu category: {missing_category}")
```

```python
# Đọc lại file link và bổ sung các dòng thiếu category
df_link = pd.read_csv(r"D:\OU\KhoaLuanTotNghiep\DrugRecommandation\crawlData_Pharmacity\pharmacity_medicine_links.csv")
df_link = df_link.copy()

df_add_category = df_link[df_link['links'].isin(df_merged['url'])][['links', 'category_name']].drop_duplicates(subset='links')
df_final = pd.merge(df_merged, df_add_category, left_on='url', right_on='links', how='left', suffixes=('', '_new'))
df_final['category_name'] = df_final['category_name'].combine_first(df_final['category_name_new'])
df_final.drop(columns=['category_name_new', 'links'], inplace=True)
df_final.to_csv("pharmacity_medicines_data_final.csv", index=False)
```

```python
df_new = pd.read_csv("pharmacity_medicines_data_final.csv")
print(df_new.shape)
df_new.head()
```

```python
# Kiểm tra còn dòng nào thiếu category không
missing_category_after = df_new['category_name'].isna().sum()
print(f"Số dòng còn thiếu category sau bổ sung: {missing_category_after}")
```
