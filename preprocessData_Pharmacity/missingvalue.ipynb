```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pharmacity_df = pd.read_csv("pharmacity_medicines_data_final.csv")
print(pharmacity_df.shape)
pharmacity_df.info()
pharmacity_df.head()
```

```python
# Lọc ra các link thuốc còn thiếu thông tin 'chỉ định' và 'chỉ định chi tiết'
indication_cols = ["chi_dinh_tom_tat", "chi_dinh_chi_tiet"]

# Chuẩn hoá giá trị thiếu
for col in indication_cols:
    pharmacity_df[col] = pharmacity_df[col].astype("string").str.strip()

missing_tokens = {"": pd.NA, "nan": pd.NA, "None": pd.NA, "null": pd.NA, "N/A": pd.NA}
pharmacity_df[indication_cols] = pharmacity_df[indication_cols].replace(missing_tokens)

# Thiếu ít nhất 1 trường chỉ định
missing_any_mask = pharmacity_df[indication_cols].isna().any(axis=1)
missing_any_df = pharmacity_df.loc[missing_any_mask, ["url", "name", *indication_cols, "mo_ta"]].copy()

# Thiếu cả 2 trường chỉ định
missing_both_mask = pharmacity_df[indication_cols].isna().all(axis=1)
missing_both_df = pharmacity_df.loc[missing_both_mask, ["url", "name", "mo_ta"]].copy()

print(f"Tổng số thuốc: {len(pharmacity_df):,}")
print(f"Thiếu >=1 trường chỉ định: {len(missing_any_df):,}")
print(f"Thiếu cả 2 trường chỉ định: {len(missing_both_df):,}")

# Lưu danh sách link để crawl/bổ sung sau
missing_any_df[["url"]].dropna().drop_duplicates().to_csv(
    "pharmacity_missing_indication_links.csv", index=False, encoding="utf-8-sig"
 )
missing_both_df[["url"]].dropna().drop_duplicates().to_csv(
    "pharmacity_missing_both_indication_links.csv", index=False, encoding="utf-8-sig"
 )

missing_any_df.head(10)
```
