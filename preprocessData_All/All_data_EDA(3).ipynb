<h3>Kiểm tra sơ bộ các bộ dữ liệu</h3>

```python
# Kiểm tra các cột
datasets = {
    "DieuTri": dieutri_df,
    "LongChau": longchau_df,
    "AnKhang": ankhang_df,
    "Pharmacity": pharmacity_df,
    "MinhChau": minhchau_df
}


for name, df in datasets.items():
    print(f"Dataset: {name}")
    print("Columns:", df.columns)
    print("Missing values:\n", df.isnull().sum())
    print("\n")
```

```python
# Tìm cột chung và cột khác nhau giữa các dataset
all_columns = [set(df.columns) for df in datasets.values()]
common_columns = set.intersection(*all_columns)
print("Các cột chung:", common_columns)


for name, df in datasets.items():
    unique_columns = set(df.columns) - common_columns
    print(f"Các cột riêng của {name}:", unique_columns)
```

```python
# Kiểm tra giá trị thiếu
for name, df in datasets.items():
    print(f"\nDataset: {name}")
    missing = df.isnull().sum()
    missing_percent = round((missing / len(df)) * 100, 2)
    missing_df = pd.DataFrame({'Missing Values': missing, 'Percentage': missing_percent})
    print(missing_df[missing_df['Missing Values'] > 0])
```

```python
# Kiểm tra dữ liệu trùng lặp
for name, df in datasets.items():
    print(f"\nDataset: {name}")
    print(f"Tổng số dòng: {len(df)}")
    print(f"Số dòng trùng lặp: {df.duplicated().sum()}")


    name_cols = [col for col in df.columns if 'name' in col.lower() or 'Tên thuốc' in col.lower()
                 or 'ten_thuoc' in col.lower()]
    if name_cols:
        print(f"Số tên thuốc trùng lặp theo cột '{name_cols[0]}': {df.duplicated(subset=[name_cols[0]]).sum()}")
```

```python
# Thống kê mô tả dữ liệu
for name, df in datasets.items():
    print(f"\nDataset: {name}")
    print(df.describe(include='all'))
```

```python
# Kiểm tra category
for name, df in datasets.items():
    category_cols = [col for col in df.columns if 'category' in col.lower() or 'category_name' in col.lower()
                     or 'Category Name' in col.lower()]
    for col in category_cols:
        print(f"\nDataset: {name} - Cột: {col}")
        print(f"Số lượng giá trị unique: {df[col].nunique()}")
        print("Top 10 giá trị: ")
        print(df[col].value_counts().head(10))
```

```python
# Kiểm tra các giá trị đặc biệt
for name, df in datasets.items():
    print(f"Dataset: {name}")
    for col in df.columns:
        print(f"\nCột: {col}")
        s_num = pd.to_numeric(df[col], errors='coerce')

        if s_num.notna().any():
            print(f"Min: {df[col].min()}")
            print(f"Max: {df[col].max()}")
            print(f"Mean: {df[col].mean()}")
            print(f"Số giá trị = 0: {(df[col] == 0).sum()}")
            print(f"Số giá trị âm(< 0): {(df[col] < 0).sum()}")
        else:
            print("Cột non-numeric, bỏ qua thống kê")
            print(f"Số giá trị unique: {df[col].nunique()}")
            print("Top 10 giá trị: ")
            print(df[col].astype(str).value_counts(dropna=False).head(10))
```

<h3>Trực quan hóa</h3>

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Số lượng records
ax1 = axes[0, 0]
sizes = [len(df) for df in datasets.values()]
ax1.bar(datasets.keys(), sizes, color=['blue', 'orange', 'green', 'red','purple'])
ax1.set_title('Số lượng records mỗi dataset')
ax1.set_ylabel('Số lượng')
ax1.tick_params(axis='x', rotation=90)


# Số lượng cột
ax2 = axes[0, 1]
num_cols = [len(df.columns) for df in datasets.values()]
ax2.bar(datasets.keys(), num_cols, color=['blue', 'orange', 'green', 'red','purple'])
ax2.set_title('Số lượng cột mỗi dataset')
ax2.set_ylabel('Số lượng cột')
ax2.tick_params(axis='x', rotation=90)


# Giá trị thiếu (theo %)
ax3 = axes[1, 0]
missing_pcts = [round((df.isnull().sum().sum() / (len(df) * len(df.columns))) * 100, 2) for df in datasets.values()]
ax3.bar(datasets.keys(), missing_pcts, color=['blue', 'orange', 'green', 'red','purple'])
ax3.set_title('Tỷ lệ missing values mỗi dataset')
ax3.set_ylabel('Tỷ lệ (%)')
ax3.tick_params(axis='x', rotation=90)


# Giá trị trùng lặp (theo %)
ax4 = axes[1, 1]
duplicate_pcts = [round((df.duplicated().sum() / len(df)) * 100, 2) for df in datasets.values()]
ax4.bar(datasets.keys(), duplicate_pcts, color=['blue', 'orange', 'green', 'red','purple'])
ax4.set_title('Tỷ lệ duplicate records mỗi dataset')
ax4.set_ylabel('Tỷ lệ (%)')
ax4.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Ma trận so sánh cấu trúc
all_unique_cols = set()
for df in datasets.values():
    all_unique_cols.update(df.columns)

comparison_matrix = pd.DataFrame(index = sorted(all_unique_cols), columns = datasets.keys())

for col in all_unique_cols:
    for name, df in datasets.items():
        comparison_matrix.loc[col, name] = 'Yes' if col in df.columns else 'No'

print("\nMa trận so sánh cấu trúc các dataset:")
print(comparison_matrix)
```
