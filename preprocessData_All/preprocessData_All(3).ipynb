```python
import re
import unicodedata


final_columns = [
    "drug_name", "active_ingredient", "dosage_form", "strength", "packing","indication",
    "usage","side_effect","contraindication", "caution","manufacturer", "category", "price",
    "unit", "url", "image_url", "source"
]
```

```python
def norm_text(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip()
    x = unicodedata.normalize('NFC', x)
    x = re.sub(r'\s+', ' ', x)
    return x if x else np.nan
```

```python
def parse_price(x):
    if pd.isna(x):
        return np.nan
    s = str(x)
    s = re.sub(r'[^\d.,]', '', s) # Giữ lại số, dấu chấm và dấu phẩy
    s = s.replace('.', '').replace(',', '.') # Loại bỏ dấu phẩy và thay thế bằng dấu chấm
    try:
        return float(s)
    except:
        return np.nan
```

```python
def standardize_column_names(df, source_name, col_map):
    out = df.rename(columns=col_map).copy()

    for c in final_columns:
        if c not in out.columns:
            out[c] = np.nan
    
    text_cols = ["drug_name", "active_ingredient", "dosage_form", "strength", "packing", "usage",
                 "manufacturer", "category", "side_effect", "caution", "indication", "contraindication", "unit", "url"]

    for c in text_cols:
        out[c] = out[c].apply(norm_text)

    out["price"] = out["price"].map(parse_price)

    out["source"] = source_name
    out = out.dropna(subset=['drug_name'])

    out["drug_name_norm"] = out["drug_name"].fillna("").astype(str).str.lower().str.strip()
    out["dosage_form_norm"] = out["dosage_form"].fillna("").astype(str).str.lower().str.strip()
    out["active_ingredient_norm"] = out["active_ingredient"].fillna("").astype(str).str.lower().str.strip()
    out["strength_norm"] = out["strength"].fillna("").astype(str).str.lower().str.strip()
    out["manufacturer_norm"] = out["manufacturer"].fillna("").astype(str).str.lower().str.strip()

    dedup_subset = ["drug_name_norm", "dosage_form_norm", "active_ingredient_norm"]
    out = out.drop_duplicates(subset=dedup_subset, keep='first')

    return out[final_columns + ["drug_name_norm", "dosage_form_norm", "active_ingredient_norm", "strength_norm", "manufacturer_norm"]]
```

```python
# Gom cột bằng fuzzywwuzzy 
from fuzzywuzzy import fuzz,process

main_catgories = [
    'Da liễu & dị ứng',
    'Tiêu hóa & gan mật',
    'Hô hấp',
    'Tim mạch & huyết áp',
    'Thần kinh & Não bộ',
    'Cơ xương khớp',
    'Giảm đau & Hạ sốt',
    'Kháng sinh & Kháng viêm',
    'Tiết niệu & Sinh dục',
    'Tai, mắt, mũi, họng',
    'Vitamin & Khoáng chất',
    'Tiểu đường',
    'Phụ nữ & Sản phụ khoa',
    'Nam giới',
    'Ung thư',
    'Khác'
]

keyword_map = {
    "Da liễu & dị ứng": ["da liễu", "dị ứng", "mụn"],
    "Tiêu hóa & gan mật": ["tiêu hóa", "gan mật", "gan", "dạ dày"],
    "Hô hấp": ["hô hấp", "cảm lạnh", "ho"],
    "Tim mạch & huyết áp": ["tim mạch", "huyết áp", "mỡ máu", "máu"],
    "Thần kinh & Não bộ": ["thần kinh", "não", "tâm thần"],
    "Cơ xương khớp": ["cơ xương", "xương khớp", "gút", "cao xoa", "miếng dán", "dầu xoa"],
    "Giảm đau & Hạ sốt": ["giảm đau", "hạ sốt", "kháng viêm"],
    "Kháng sinh & Kháng viêm": ["kháng sinh", "kháng nấm", "viêm nhiễm"],
    "Tiết niệu & Sinh dục": ["tiết niệu", "sinh dục", "nội tiết"],
    "Tai, mắt, mũi, họng": ["mắt", "tai", "mũi", "họng", "xịt mũi", "nhỏ mắt"],
    "Vitamin & Khoáng chất": ["vitamin", "khoáng chất", "thuốc bổ"],
    "Tiểu đường": ["tiểu đường"],
    "Phụ nữ & Sản phụ khoa": ["phụ nữ", "ngừa thai", "vệ sinh phụ nữ", "sản phụ khoa","nữ","thai"],
    "Nam giới": ["nam giới", "cho nam", "nam"]
}
```

```python
def map_category(cat, fuzzy_threshold=65):
    cat_norm = norm_text(cat)

    best_match = process.extractOne(cat_norm, main_catgories, scorer=fuzz.token_set_ratio)
    if best_match and best_match[1] >= fuzzy_threshold:
        return best_match[0], best_match[1], "fuzzy"
    
    for target, kws in keyword_map.items():
        if any(kw in cat_norm for kw in kws):
            return target, best_match[1] if best_match else 0, "keyword"
    return "Khác", best_match[1] if best_match else 0, "other"
```

```python
all_categories = merged_df['category'].dropna().unique()

auto_mapping = {}
log_row  = []

for cat in all_categories:
    mapped_cat, score, method = map_category(cat, fuzzy_threshold=65)
    auto_mapping[cat] = mapped_cat
    log_row.append({"Category": cat, "Mapped Category": mapped_cat, "Score": score, "Method": method})
    print(f"Mapping '{cat}' to '{mapped_cat}' with score {score} using method {method}")
```

```python
merged_df['category_grouped'] = merged_df['category'].map(auto_mapping)
log_df = pd.DataFrame(log_row).sort_values(["Method", "Score"], ascending=[True, False])
print("Số lượng thuốc theo category đã gom nhóm:")
print(merged_df['category_grouped'].value_counts())
```
