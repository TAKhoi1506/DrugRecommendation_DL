<h3>Bổ sung các category còn thiếu</h3>

```python
# Thống kê category name trong bộ dữ liệu
category_counts = longchau_df['Category Name'].value_counts()
print(f"Tổng số lượng category: {category_counts.shape[0]}")
print(f"\nDanh sách category và số lượng thuốc tương ứng:\n{category_counts}")

# Thống kê category name theo phần trăm
category_percentages = longchau_df['Category Name'].value_counts(normalize=True) * 100
print(f"\nDanh sách category và phần trăm thuốc tương ứng:\n{category_percentages}")
```

```python
# Chuẩn hóa tên category bằng cách Loại bỏ nội dung sau ký tự '\n' 
longchau_df['Category Name'] = longchau_df['Category Name'].str.split('\n').str[0]
longchau_df['Category Name'] = longchau_df['Category Name'].str.strip()
# Thống kê lại category name sau khi chuẩn hóa
category_counts_normalized = longchau_df['Category Name'].value_counts()
print(f"Tổng số lượng category sau khi chuẩn hóa: {category_counts_normalized.shape[0]}")
print(f"\nDanh sách category và số lượng thuốc tương ứng sau khi chuẩn hóa:\n{category_counts_normalized}")
```

```python
# Lấy 10-20 tên thuốc cho mỗi category
for category in category_counts_normalized.index:
    if pd.notna(category):
        print("\n")
        print(f"\nCategory: {category}")
        print(f"Số lượng thuốc trong category: {category_counts_normalized[category]}")
        print("\n")
        
        sample_names = longchau_df[longchau_df['Category Name'] == category]['ten_thuoc'].head(15)
        for idx, name in enumerate(sample_names, 1):
            print(f"{idx}. {name}")
```

```python
# Phân tích từ khóa phổ biến trong từng category
from collections import Counter
import re

def extract_keywords(text):
    if pd.isna(text):
        return []
    
    text = re.sub(r'[^\w\s]', '', text.lower())
    words = text.split()
    return words
```

```python
# Từ khóa phổ biến trong từng category
category_keywords = {}

for category in category_counts_normalized.index:
    if pd.notna(category):
        names = longchau_df[longchau_df['Category Name'] == category]['ten_thuoc']
        all_keywords = []
        for name in names:
            keywords = extract_keywords(name)
            all_keywords.extend(keywords)

        keyword_counts = Counter(all_keywords)

        top_words = [word for word, count in keyword_counts.most_common(30) if len(word) > 2][:20]
        category_keywords[category] = top_words

        print(f"\nCategory: {category}")
        print(f"Từ khóa phổ biến: {', '.join(top_words)}")


```

```python
# Phân loại dựa trên từ khóa đã phân tích
def classify_category(name, category_keywords):
    if pd.isna(name):
        return None, 0
    
    name_lower = name.lower()
    name_keywords = extract_keywords(name_lower)
    
    # Tính điểm cho mỗi category
    scores = {}
    for category, keywords in category_keywords.items():
        score = sum(1 for keyword in keywords if keyword in name_keywords)
        scores[category] = score

    # Chọn category có điểm cao nhất
    if max(scores.values()) > 0:
        best_category = max(scores, key=scores.get)
        return best_category, scores[best_category]
    else:
        return None, 0
```

```python
# Test
for idx, row in missing_category.head(10).iterrows():
    name = row['ten_thuoc']
    predicted_category, score = classify_category(name, category_keywords)
    print(f"Thuốc: {name}")
    print(f"Dự đoán category: {predicted_category} (Score: {score})")
    print("\n")
```

```python
# Áp dụng phân loại cho toàn bộ missing_category
results = []

for idx, row in missing_category.iterrows():
    name = row['ten_thuoc']
    url = row['url']
    predicted_category, score = classify_category(name, category_keywords)
    results.append(
        {
            'url': url,
            'ten_thuoc': name,
            'predicted_category': predicted_category,
            'score': score
        }
    )
results_df = pd.DataFrame(results)
results_df
```

```python
# Thống kê theo category dự đoán
print("Thống kê số lượng thuốc theo category dự đoán:")
predicted_category_counts = results_df['predicted_category'].value_counts()
print(predicted_category_counts)

# Kiểm tra các trường hợp không dự đoán được category
no_match = results_df[results_df['score'] == 0]
print(f"Số lượng thuốc không dự đoán được category: {len(no_match)}")
if len(no_match) > 0:
    print(no_match['ten_thuoc', 'predicted_category', 'score'])
    
```

```python
# Cập nhật vào dataframe gốc
for idx, row in results_df.iterrows():
    if row['predicted_category'] is not None:
        longchau_df.loc[longchau_df['url'] == row['url'], 'Category Name'] = row['predicted_category']


# Kiểm tra lại số lượng category sau khi cập nhật
print(f"Số dòng còn thiếu category sau khi cập nhật: {longchau_df['Category Name'].isnull().sum()}")

# Thống kê lại category name sau khi cập nhật
category_counts_updated = longchau_df['Category Name'].value_counts()
print(f"Tổng số lượng category sau khi cập nhật: {category_counts_updated}")
```
