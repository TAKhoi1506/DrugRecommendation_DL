```python
import pandas as pd
import numpy as np
import re
```

```python
longchau_df = pd.read_csv("Longchau_data_with_category_updated.csv", encoding='utf-8-sig')
longchau_df.info()
longchau_df.head()
```

```python

def clean_list_format(text):
    if pd.isna(text):
        return text
    
    text = str(text)
    text = text.replace('[', '').replace(']', '') 
    text = text.replace("'", '')
    text = text.strip()
    return text


columns_to_clean = ['thanh_phan', 'cong_dung', 'chong_chi_dinh', 'luu_y' ,'lieu_dung', 'tac_dung_phu', 'bao_quan', 'mo_ta']

for col in columns_to_clean:
    if col in longchau_df.columns:
        longchau_df[col] = longchau_df[col].apply(clean_list_format)

longchau_df.to_csv("Longchau_data_with_category_cleaned.csv", index=False, encoding='utf-8-sig')

print("\n=== THANH PHAN ===")
print(longchau_df['thanh_phan'].iloc[0])
print("\n=== CONG_DUNG ===")
print(longchau_df['cong_dung'].iloc[0])
```
