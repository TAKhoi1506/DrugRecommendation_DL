```python
import pandas as pd
import numpy as np
import re
```

```python
# Đọc file Long Châu đã xử lý
df_preprocessed = pd.read_csv("Longchau_data_processed.csv")
df_preprocessed.shape
df_preprocessed
```

```python
# Đọc file Category mới cào thêm
df_category = pd.read_csv(r"D:\OU\KhoaLuanTotNghiep\DrugRecommandation\crawlData_LongChau\longchau_medicine_category_links(3).csv")
df_category.shape
df_category.head()
```

```python
df_combined = df_preprocessed.merge(
    df_category[['Medicine Link', 'Category Name', 'Category Link']], 
    left_on='url',  # tên cột trong df_preprocessed (thay 'url' nếu khác)
    right_on='Medicine Link',  # tên cột trong df_category
    how='left'
)

missing = df_combined['Category Name'].isna().sum()
print(f"Số lượng thuốc không tìm thấy category: {missing}")

df_combined.to_csv("Longchau_data_with_category.csv", index=False)
```
